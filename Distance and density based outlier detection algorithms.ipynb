```python
from google.colab import drive
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = '/content/AirPassengers.csv'
df = pd.read_csv(file_path)

df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')
```

```python
df.set_index('date', inplace=True)

df.head()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['value'])
plt.title('Number of Air Passengers Over Time')
plt.xlabel('Time')
plt.ylabel('Number of Passengers')
plt.grid(True)
plt.show()
```

```python
df['Year'] = df.index.year
df['Month'] = df.index.month

plt.figure(figsize=(12, 6))
sns.lineplot(x='Month', y='value', hue='Year', data=df, palette='tab10')
plt.title('Seasonal Plot of Air Passengers')
plt.xlabel('Month')
plt.ylabel('Number of Passengers')
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
# Seasonal Subseries Plot
fig, axes = plt.subplots(3, 4, figsize=(15, 10), sharey=True)
fig.suptitle('Seasonal Subseries Plot of Air Passengers', fontsize=16)

for i, ax in enumerate(axes.flatten(), start=1):
    sns.lineplot(x='Year', y='value', data=df[df['Month'] == i], ax=ax)
    ax.set_title(f'Month: {i}')
    ax.set_xlabel('Year')
    ax.set_ylabel('Passengers')
    ax.grid(True)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
